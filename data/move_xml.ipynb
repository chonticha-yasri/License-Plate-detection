```python
import os 
import shutil
```

```python
os.mkdir('')
```

```python
fol = ['train','test', 'val']
for f in fol:
    os.mkdir(f'{f}/annotations')
    for files in os.listdir(f'{f}/labels/'):
        print(f"copy annotations/{files.strip('.txt')}.xml to {f}/annotations/{files.strip('.txt')}.xml")
        shutil.copyfile(f"annotations/{files.strip('.txt')}.xml", f"{f}/annotations/{files.strip('.txt')}.xml")
```
